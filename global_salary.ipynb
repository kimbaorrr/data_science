```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import random
import matplotlib.pyplot as plt
```

```python
# Nạp dữ liệu CSV
df = pd.read_csv('/content/drive/MyDrive/TF_Training/datasets/data_science/global_salary.csv', delimiter=',')
df.sample(50)
```

```python
# Kiểm tra giá trị N/A
df.isna().sum()
```

```python
# Kiểm tra phần tử ngoại lai
df.describe()
```

```python
px.pie(
    df,
    names='continent_name',
    values='highest_salary',
    color='continent_name',
    title='Thống kê mức thu nhập cao nhất theo từng châu lục'
    ).show()
# => EU có thu nhập cao nhất
```

```python
top_10 = df.sort_values('highest_salary', ascending=False).head(10)
px.bar(
    top_10,
    x='country_name',
    y='highest_salary',
    title='Top 10 quốc gia có mức thu nhập cao nhất',
    color='highest_salary'
    ).show()
# Switchland có mức thu nhập cao nhất
```

```python
top_10 = df.sort_values('lowest_salary', ascending=True).head(10)
px.bar(
    top_10,
    x='country_name',
    y='lowest_salary',
    title='Top 5 quốc gia có mức thu nhập thấp nhất',
    color='country_name'
    ).show()
# Zambia có thu nhập thấp nhất
```

```python
top_10 = df.sort_values('average_salary', ascending=False).head(10)
px.bar(
    top_10,
    x='country_name',
    y='average_salary',
    title='Top 10 quốc gia có mức thu nhập trung bình cao nhất',
    color='average_salary'
    ).show()
```

```python
top_10 = df[df['continent_name'] == 'Asia'].sort_values('highest_salary', ascending=False).head(10)
px.bar(
    top_10,
    x='country_name',
    y='highest_salary',
    color='country_name'
).show()
# Singapore là nước có mức thu nhập cao nhất ở Asia
```

```python
top_10 = df[df['continent_name'] == 'Asia'].sort_values('lowest_salary', ascending=True).head(10)
px.bar(
    top_10,
    x='country_name',
    y='lowest_salary',
    color='country_name'
).show()
# Syria là nước có mức thu nhập thấp nhất ở Asia
```
