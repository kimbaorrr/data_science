```python
import matplotlib.pyplot as plt
import pandas as pd
from classify.common import random_color_str
import seaborn as sns
import plotly.express as px
```

```python
sns.set_style('whitegrid')
```

```python
# Nạp dữ liệu CSV
df = pd.read_csv('datasets/global_salary/salary_data.csv', delimiter=',')
df.sample(50)
```

```python
df.columns
```

```python
# In thông tin thuộc tính
df.info()
```

```python
# Kiểm tra giá trị N/A
df.isna().sum()
```

```python
# Tóm tắt thuộc tính
df.describe()
```

```python
a = px.pie(df, names='continent_name', values='highest_salary', title='Thống kê mức thu nhập cao nhất theo từng châu lục')
a.show()
```

```python
top_10 = df.sort_values('highest_salary', ascending=False).head(10)
a = px.bar(top_10, x='country_name', y='highest_salary', title='Top 10 quốc gia có mức thu nhập cao nhất', color='highest_salary')
a.show()
```

```python
top_5 = df.sort_values('lowest_salary', ascending=True).head(5)
a = px.bar(top_5, x='country_name', y='lowest_salary', title='Top 5 quốc gia có mức thu nhập thấp nhất', color='country_name')
a.show()
```

```python
top_10 = df.sort_values('median_salary', ascending=False).head(10)
a = px.bar(top_10, x='country_name', y='median_salary', title='Top 10 quốc gia có mức thu nhập trung vị cao nhất', color='median_salary')
a.show()
```
